```python
with open('/kaggle/input/d/ahmedabelal/arabic-poetry/Arabic_poetry_dataset.csv','r') as f:
    poems1=pd.read_csv(f)
```

```python
with open('/kaggle/input/arabic-poetry-dataset/Arabic_Poetry_Dataset.csv','r') as f:
    poems2=pd.read_csv(f)
```

```python
with open('/kaggle/input/arabic-poetry-dataset-478-2017/all_poems.csv','r') as f:
    poems3=pd.read_csv(f)
```

```python
unique_poems=set()
poem_text=[]
all_poems=[poems1,poems2,poems3]

for df in all_poems:
    mask = ~df['poem_title'].isin(unique_poems)    
    unique_poems.update(df.loc[mask, 'poem_title'])    
    poem_text.extend(df.loc[mask, 'poem_text'].tolist())
```

```python
poem_text = [p for p in poem_text if isinstance(p, str)]
```

```python
text='\n'.join(poem_text)
```

```python
import re
arabic_letters=re.compile('[\u061F-\u064A]|[\u0671-\u06D5]|\s+|\n') 
```

```python
def remove_non_matching(text,pattern):
    return(''.join(re.findall(pattern,text)))
```

```python
new_text=remove_non_matching(text,arabic_letters)
```
